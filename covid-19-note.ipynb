## Kmeans

```python
from sklearn.cluster import KMeans
```

```python
distortions = []
K = range(1,15)
for k in K:
    kmeanModel = KMeans(n_clusters=k,n_jobs=-1)
    kmeanModel.fit(df_std.loc[df_std["cluster"]==1])
    distortions.append(kmeanModel.inertia_)
```

```python
plt.figure(figsize=(16,8))
plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The Elbow Method showing the optimal k')
plt.show()
```

## NYC subway stations

```python
sub = pd.read_csv("DOITT_SUBWAY_STATION_01_13SEPT2010.csv")[["NAME","the_geom"]]
sub.head()
```

```python
sub["latitude"]=sub["the_geom"].apply(lambda x: x.strip(")").replace("(","").split(" ")[2]).astype("float")
sub["longitude"]=sub["the_geom"].apply(lambda x: x.strip(")").replace("(","").split(" ")[1]).astype("float")
```

```python
from uszipcode import SearchEngine
```

```python
search= SearchEngine()
```

```python
sub["zipcode"]=0
```

```python
for i in range(len(sub)):
    sub["zipcode"][i]= search.by_coordinates(sub["latitude"][i],sub["longitude"][i],radius=5)[0].zipcode
```

```python
sub.zipcode=sub.zipcode.astype("int")
```

```python
sub = sub.groupby("zipcode").count().reset_index()[["zipcode","NAME"]]
```

```python
sub.columns= ["zipcode","subway_count"]
```

```python
sub.info
```

```python
df= df.merge(sub,left_on="geo_id",right_on="zipcode",how="left")
```

```python
df.drop("do_date",axis=1,inplace=True)
```

```python
df.subway_count = df["subway_count"].fillna(0)
```

```python
df.to_csv("complete_nyc.csv",index=False)
```
